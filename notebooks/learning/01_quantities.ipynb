## Goals for Today

## Why this matters for radio astronomy
- Radio frequencies (MHz, GHz) need conversion to wavelength (meters)
- Flux densities (Jy) to physical power (W/Hz)
- Baseline lengths (meters) to wavelengths for interferometry

## Checkpoint (Day 1)
- 10 light-years converted to meters and parsecs
- 5 parsecs converted to light-years and AU
- A calculation that demonstrates unit mismatch (e.g., 10 * u.km + 5 * u.s) raises an error

```python
import astropy.units as u
```

```python
# Converting 10 light-years to meters and parsecs

distance_lyr = 10 * u.lyr
print(f"Distance in light years: {distance_lyr}")
distance_m = distance_lyr.to(u.m)
print(f"Distance in meters: {distance_m}")
distance_pc = distance_m.to(u.pc)
print(f"Distance in parsecs: {distance_pc}")
```

```python
# Converting 5 parsecs to light-years
distance_pc = 5 * u.pc
print(f"Distance in parsecs: {distance_pc}")
distance_lyr = distance_pc.to(u.lyr)
print(f"Distance in light years: {distance_lyr}")
```

```python
# Calculation demonstrating unit mismatch 

print(20 * u.km + 0.5 * u.pc)  # demonstrate same physical quantity
print(20 * u.m + 8 * u.s) # demonstrate unit mismatch
```
