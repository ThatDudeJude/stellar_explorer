# Stellar Coordinate Explorer - Sky Coordinates

## Objectives
Define celestial coordinates in different systems (ICRS, Galactic) and transform between them.

## Why this maters for radio astronomy
- Pointing a radio telescope requires converting catalogue coordinates (RA/Dec) to local horizontal coordinates (alt/az)
- Galactic coordinates help interpret emission from the Milky Way
- Interferometric calibration uses known bright sources (e.g., 3C286) with precise ICRS positions

## Dataset
No catalogue yet - we work with a single star: **Sitius** ($\alpha$ CMa)
- ICRS coordinates (from Gaia):
  - RA = $101.287^\degree$
  - Dec = $-16.716^\degree$

## Goals for Today
- Create `SkyCoord` object for Sirius in ICRS
- Transform to Galactic coordinates
- Print both ICRS and Galactic representations
- (Optional) Transform to FKS or other frames

## Checkpoint (Day 2)
- Sirius ICRS defined and printed
- Galactic coordinates computed (l, b)
- Transform back from Galactic to ICRS and verify consistency

```python
# imports
from astropy.coordinates import SkyCoord, Galactic
import astropy.coordinates as coord
import astropy.units as u
```

```python
# Equitorial coordinate frame
sirius_icrs = SkyCoord(101.287, - 16.716, unit=(u.deg, u.deg), frame="icrs")
print(sirius_icrs)
print(f"Sirius RA: {sirius_icrs.ra}")
print(f"Sirius Dec: {sirius_icrs.dec}")
```

```python
# Transform to Galactic coordinate frame
sirius_galactic = sirius_icrs.transform_to(Galactic())
print(sirius_galactic)
print(f"Sirius galactic latitude b: {sirius_galactic.b}")
print(f"Sirius galactic longitude l: {sirius_galactic.l}")
```

```python
# Transform back to Equitorial coordinate frame
sirius_icrs = sirius_galactic.transform_to(coord.ICRS())
print(sirius_icrs)
print(f"Sirius RA: {sirius_icrs.ra}")
print(f"Sirius Dec: {sirius_icrs.dec}")
```
